```python
import multiprocessing as mp
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import numpy as np
count = 1000000
mu, sigma = 0, 2/3
a = np.random.normal(mu, sigma, 3 * count)
np.clip(a, -1, 1, out=a)
a = a.reshape(count, 3)
b = a.sum(axis=1) / 3
a.shape, b.shape, a[0], b[0]

# plot
sns.set_style('ticks')
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)
sns.distplot(b);
```
