```python
from google_images_download import google_images_download
from multiprocessing import Pool

def load_images(keyword: str):
    response = google_images_download.googleimagesdownload()
    arguments = {
        "keywords": keyword, 
        "size": ">400*300",
        "format": "jpg",
        "limit": 5, 
        "output_directory": "images",
        "no_directory": True
    }
    response.download(arguments)

prefixes = 'chinese,thai,japanese,'.split(',')
words = 'journal,news,newspaper,newspapers,letter,' \
        'letters,alphabet,text,texts,sting,stings'.split(',')
    
keywords = [
    '%s %s' % (prefix, word) 
    for prefix in prefixes 
    for word in words
]
with Pool(processes=20) as pool:
    pool.map(load_images, keywords)
```
